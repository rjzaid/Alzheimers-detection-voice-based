```python
# ============================================================
# FINE-TUNE Data2Vec Audio Model (partial unfreezing)
# ============================================================

import os
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, random_split
import librosa
import numpy as np
import pandas as pd
from transformers import AutoFeatureExtractor, Data2VecAudioModel
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix

# ============================================================
# 1️⃣ Dataset Class — Load Audio + Label
# ============================================================
class AlzheimerAudioDataset(Dataset):
    def __init__(self, csv_path, audio_dir, extractor, max_length=160000):
        self.data = pd.read_csv(csv_path)
        self.audio_dir = audio_dir
        self.extractor = extractor
        self.max_length = max_length

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        row = self.data.iloc[idx]
        file_path = os.path.join(self.audio_dir, row['file'])
        label = torch.tensor(row['label'], dtype=torch.float32).unsqueeze(0)
        
        # Load audio
        speech_array, sr = librosa.load(file_path, sr=16000)
        
        # Feature extraction
        inputs = self.extractor(
            speech_array,
            sampling_rate=sr,
            return_tensors="pt",
            padding="max_length",
            truncation=True,
            max_length=self.max_length
        )
        return inputs.input_values.squeeze(0), label


# ============================================================
# 2️⃣ Model — Data2Vec + Classifier Head
# ============================================================
class Data2VecClassifier(nn.Module):
    def __init__(self, base_model_name="facebook/data2vec-audio-base-960h"):
        super().__init__()
        self.data2vec = Data2VecAudioModel.from_pretrained(base_model_name)
        
        # 🔒 Freeze feature extractor
        for param in self.data2vec.feature_extractor.parameters():
            param.requires_grad = False
        
        # 🔒 Freeze lower encoder layers (0–8)
        for i, layer in enumerate(self.data2vec.encoder.layers):
            if i < 9:
                for param in layer.parameters():
                    param.requires_grad = False
        
        # ✅ Unfreeze last 3 encoder layers
        for i, layer in enumerate(self.data2vec.encoder.layers):
            if i >= 9:
                for param in layer.parameters():
                    param.requires_grad = True

        # 🧠 Classifier head
        self.classifier = nn.Sequential(
            nn.Linear(768, 300),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(300, 1),
            nn.Sigmoid()
        )

    def forward(self, input_values):
        outputs = self.data2vec(input_values)
        hidden_states = outputs.last_hidden_state
        pooled = hidden_states.mean(dim=1)
        logits = self.classifier(pooled)
        return logits


# ============================================================
# 3️⃣ Training + Evaluation
# ============================================================
def train_model(model, train_loader, val_loader, device, epochs=50, lr=1e-3):
    model.to(device)
    criterion = nn.BCELoss()
    optimizer = torch.optim.Adam(filter(lambda p: p.requires_grad, model.parameters()), lr=lr)

    train_losses, val_losses = [], []

    for epoch in range(epochs):
        model.train()
        total_train_loss = 0

        for x_batch, y_batch in train_loader:
            x_batch, y_batch = x_batch.to(device), y_batch.to(device)
            optimizer.zero_grad()
            outputs = model(x_batch)
            loss = criterion(outputs, y_batch)
            loss.backward()
            optimizer.step()
            total_train_loss += loss.item() * x_batch.size(0)

        avg_train_loss = total_train_loss / len(train_loader.dataset)
        train_losses.append(avg_train_loss)

        # Validation
        model.eval()
        total_val_loss = 0
        with torch.no_grad():
            for x_batch, y_batch in val_loader:
                x_batch, y_batch = x_batch.to(device), y_batch.to(device)
                outputs = model(x_batch)
                loss = criterion(outputs, y_batch)
                total_val_loss += loss.item() * x_batch.size(0)

        avg_val_loss = total_val_loss / len(val_loader.dataset)
        val_losses.append(avg_val_loss)

        print(f"Epoch {epoch+1}/{epochs} | Train Loss: {avg_train_loss:.4f} | Val Loss: {avg_val_loss:.4f}")

    # Plot
    plt.figure(figsize=(7, 5))
    plt.plot(train_losses, label='Train Loss')
    plt.plot(val_losses, label='Validation Loss')
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()
    plt.title("Data2Vec Fine-Tuning Loss")
    plt.show()

    return model


# ============================================================
# 4️⃣ Test Evaluation Function
# ============================================================
def evaluate_model(model, test_loader, device):
    model.eval()
    y_true, y_pred = [], []
    with torch.no_grad():
        for x_batch, y_batch in test_loader:
            x_batch = x_batch.to(device)
            outputs = model(x_batch)
            preds = (outputs.cpu().numpy() > 0.5).astype(int)
            y_pred.extend(preds)
            y_true.extend(y_batch.cpu().numpy())
    
    acc = accuracy_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    print(f"✅ Test Accuracy: {acc:.4f} | F1 Score: {f1:.4f}")
    print("Confusion Matrix:\n", confusion_matrix(y_true, y_pred))


# ============================================================
# 5️⃣ Prepare Data (Train / Val / Test Split)
# ============================================================
csv_path = "/kaggle/input/alzhiemer/DATASET/label.csv"
audio_dir = "/kaggle/input/alzhiemer/DATASET"
extractor = AutoFeatureExtractor.from_pretrained("facebook/data2vec-audio-base-960h")

dataset = AlzheimerAudioDataset(csv_path, audio_dir, extractor)

# 80% train, 10% val, 10% test
train_size = int(0.8 * len(dataset))
val_size = int(0.1 * len(dataset))
test_size = len(dataset) - train_size - val_size
train_ds, val_ds, test_ds = random_split(dataset, [train_size, val_size, test_size])

train_loader = DataLoader(train_ds, batch_size=4, shuffle=True)
val_loader = DataLoader(val_ds, batch_size=4, shuffle=False)
test_loader = DataLoader(test_ds, batch_size=4, shuffle=False)

# ============================================================
# 6️⃣ Train + Evaluate
# ============================================================
device = "cuda" if torch.cuda.is_available() else "cpu"
model = Data2VecClassifier()

# Training
model = train_model(model, train_loader, val_loader, device=device, epochs=25, lr=1e-5)

# Final evaluation on test set
evaluate_model(model, test_loader, device)
```

```python
# ============================================================
# FINE-TUNE Data2Vec Audio Model (with Train vs Val Accuracy Plot)
# ============================================================

import os
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, random_split
import librosa
import numpy as np
import pandas as pd
from transformers import AutoFeatureExtractor, Data2VecAudioModel
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix

# ============================================================
# 1️⃣ Dataset Class — Load Audio + Label
# ============================================================
class AlzheimerAudioDataset(Dataset):
    def __init__(self, csv_path, audio_dir, extractor, max_length=160000):
        self.data = pd.read_csv(csv_path)
        self.audio_dir = audio_dir
        self.extractor = extractor
        self.max_length = max_length

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        row = self.data.iloc[idx]
        file_path = os.path.join(self.audio_dir, row['file'])
        label = torch.tensor(row['label'], dtype=torch.float32).unsqueeze(0)
        
        speech_array, sr = librosa.load(file_path, sr=16000)
        inputs = self.extractor(
            speech_array,
            sampling_rate=sr,
            return_tensors="pt",
            padding="max_length",
            truncation=True,
            max_length=self.max_length
        )
        return inputs.input_values.squeeze(0), label


# ============================================================
# 2️⃣ Model — Data2Vec + Classifier Head
# ============================================================
class Data2VecClassifier(nn.Module):
    def __init__(self, base_model_name="facebook/data2vec-audio-base-960h"):
        super().__init__()
        self.data2vec = Data2VecAudioModel.from_pretrained(base_model_name)

        # Freeze feature extractor
        for param in self.data2vec.feature_extractor.parameters():
            param.requires_grad = False

        # Freeze first 9 layers, unfreeze last 3
        for i, layer in enumerate(self.data2vec.encoder.layers):
            if i < 9:
                for param in layer.parameters():
                    param.requires_grad = False
            else:
                for param in layer.parameters():
                    param.requires_grad = True

        self.classifier = nn.Sequential(
            nn.Linear(768, 300),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(300, 1),
            nn.Sigmoid()
        )

    def forward(self, input_values):
        outputs = self.data2vec(input_values)
        hidden_states = outputs.last_hidden_state
        pooled = hidden_states.mean(dim=1)
        logits = self.classifier(pooled)
        return logits


# ============================================================
# 3️⃣ Training + Evaluation (with accuracy tracking)
# ============================================================
def train_model(model, train_loader, val_loader, device, epochs=15, lr=1e-5):
    model.to(device)
    criterion = nn.BCELoss()
    optimizer = torch.optim.Adam(filter(lambda p: p.requires_grad, model.parameters()), lr=lr)

    train_losses, val_losses = [], []
    train_accs, val_accs = [], []

    for epoch in range(epochs):
        # ----------------- TRAIN -----------------
        model.train()
        total_train_loss, correct_train, total_train = 0, 0, 0

        for x_batch, y_batch in train_loader:
            x_batch, y_batch = x_batch.to(device), y_batch.to(device)
            optimizer.zero_grad()
            outputs = model(x_batch)
            loss = criterion(outputs, y_batch)
            loss.backward()
            optimizer.step()

            total_train_loss += loss.item() * x_batch.size(0)
            preds = (outputs > 0.5).float()
            correct_train += (preds == y_batch).sum().item()
            total_train += y_batch.size(0)

        avg_train_loss = total_train_loss / len(train_loader.dataset)
        avg_train_acc = correct_train / total_train

        # ----------------- VALIDATION -----------------
        model.eval()
        total_val_loss, correct_val, total_val = 0, 0, 0
        with torch.no_grad():
            for x_batch, y_batch in val_loader:
                x_batch, y_batch = x_batch.to(device), y_batch.to(device)
                outputs = model(x_batch)
                loss = criterion(outputs, y_batch)
                total_val_loss += loss.item() * x_batch.size(0)
                preds = (outputs > 0.5).float()
                correct_val += (preds == y_batch).sum().item()
                total_val += y_batch.size(0)

        avg_val_loss = total_val_loss / len(val_loader.dataset)
        avg_val_acc = correct_val / total_val

        # Save results
        train_losses.append(avg_train_loss)
        val_losses.append(avg_val_loss)
        train_accs.append(avg_train_acc)
        val_accs.append(avg_val_acc)

        print(f"Epoch {epoch+1}/{epochs} | "
              f"Train Loss: {avg_train_loss:.4f} | Val Loss: {avg_val_loss:.4f} | "
              f"Train Acc: {avg_train_acc*100:.2f}% | Val Acc: {avg_val_acc*100:.2f}%")

    # ----------------- PLOTS -----------------
    plt.figure(figsize=(12,5))

    # Loss plot
    plt.subplot(1, 2, 1)
    plt.plot(train_losses, label='Train Loss')
    plt.plot(val_losses, label='Validation Loss')
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.title("Training vs Validation Loss")
    plt.legend()

    # Accuracy plot
    plt.subplot(1, 2, 2)
    plt.plot(train_accs, label='Train Accuracy')
    plt.plot(val_accs, label='Validation Accuracy')
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.title("Training vs Validation Accuracy")
    plt.legend()

    plt.tight_layout()
    plt.show()

    return model


# ============================================================
# 4️⃣ Test Evaluation
# ============================================================
def evaluate_model(model, test_loader, device):
    model.eval()
    y_true, y_pred = [], []
    with torch.no_grad():
        for x_batch, y_batch in test_loader:
            x_batch = x_batch.to(device)
            outputs = model(x_batch)
            preds = (outputs.cpu().numpy() > 0.5).astype(int)
            y_pred.extend(preds)
            y_true.extend(y_batch.cpu().numpy())

    acc = accuracy_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    print(f"\n✅ Test Accuracy: {acc:.4f} | F1 Score: {f1:.4f}")
    print("Confusion Matrix:\n", confusion_matrix(y_true, y_pred))


# ============================================================
# 5️⃣ Data Split
# ============================================================
csv_path = "/kaggle/input/alzhiemer/DATASET/label.csv"
audio_dir = "/kaggle/input/alzhiemer/DATASET"
extractor = AutoFeatureExtractor.from_pretrained("facebook/data2vec-audio-base-960h")

dataset = AlzheimerAudioDataset(csv_path, audio_dir, extractor)
train_size = int(0.8 * len(dataset))
val_size = int(0.1 * len(dataset))
test_size = len(dataset) - train_size - val_size
train_ds, val_ds, test_ds = random_split(dataset, [train_size, val_size, test_size])

train_loader = DataLoader(train_ds, batch_size=4, shuffle=True)
val_loader = DataLoader(val_ds, batch_size=4, shuffle=False)
test_loader = DataLoader(test_ds, batch_size=4, shuffle=False)


# ============================================================
# 6️⃣ Train + Evaluate
# ============================================================
device = "cuda" if torch.cuda.is_available() else "cpu"
model = Data2VecClassifier()

# Train with accuracy tracking
model = train_model(model, train_loader, val_loader, device=device, epochs=15, lr=1e-5)

# Final evaluation
evaluate_model(model, test_loader, device)
```

```python
# ============================================================
# 7️⃣ Save All Embeddings in a Folder
# ============================================================
def save_all_embeddings(model, dataset, device, save_dir="final_embeddings"):
    os.makedirs(save_dir, exist_ok=True)
    model.eval()
    extractor = dataset.extractor

    with torch.no_grad():
        for idx in range(len(dataset)):
            x, y = dataset[idx]
            x = x.unsqueeze(0).to(device)
            outputs = model.data2vec(x)
            embedding = outputs.last_hidden_state.mean(dim=1).cpu().numpy()  # pooled embedding

            file_name = dataset.data.iloc[idx]['file'].replace('.wav', '.npy')
            save_path = os.path.join(save_dir, file_name)
            np.save(save_path, embedding)
    
    print(f"✅ All embeddings saved in folder: {save_dir}")

# ============================================================
# Save embeddings after training
# ============================================================
save_all_embeddings(model, dataset, device, save_dir="final_embeddings")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# -----------------------------
# 1️⃣ Load matched CSV and embeddings
# -----------------------------
csv_path = "matched_embeddings_mmse.csv"
embeddings_dir = "final_embeddings"
df = pd.read_csv(csv_path)

# Custom Dataset
class EmbeddingDataset(Dataset):
    def __init__(self, df, embeddings_dir):
        self.df = df
        self.embeddings_dir = embeddings_dir

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        embedding = np.load(os.path.join(self.embeddings_dir, row['file_name']))
        embedding = torch.tensor(embedding, dtype=torch.float32).squeeze()
        mmse = torch.tensor(row['mmse_score'], dtype=torch.float32)
        return embedding, mmse

dataset = EmbeddingDataset(df, embeddings_dir)

# -----------------------------
# 2️⃣ Train / Val / Test Split
# -----------------------------
total = len(dataset)
train_size = int(0.8 * total)
val_size = int(0.1 * total)
test_size = total - train_size - val_size

train_set, val_set, test_set = random_split(dataset, [train_size, val_size, test_size])
train_loader = DataLoader(train_set, batch_size=16, shuffle=True)
val_loader = DataLoader(val_set, batch_size=16)
test_loader = DataLoader(test_set, batch_size=16)

# -----------------------------
# 3️⃣ Neural Network
# -----------------------------
class ADSeverityNN(nn.Module):
    def __init__(self, input_dim=768):
        super(ADSeverityNN, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(input_dim, 100),
            nn.ReLU(),
            nn.Linear(100, 200),
            nn.ReLU(),
            nn.Linear(200, 1)
        )

    def forward(self, x):
        return self.model(x)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = ADSeverityNN(input_dim=768).to(device)

# Loss and optimizer
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)

# -----------------------------
# 4️⃣ Training Loop with Loss History
# -----------------------------
num_epochs = 50
train_losses, val_losses = [], []

for epoch in range(num_epochs):
    # Training
    model.train()
    train_loss = 0
    for embeddings, mmse in train_loader:
        embeddings, mmse = embeddings.to(device), mmse.to(device)
        optimizer.zero_grad()
        outputs = model(embeddings).squeeze()
        loss = criterion(outputs, mmse)
        loss.backward()
        optimizer.step()
        train_loss += loss.item() * embeddings.size(0)
    train_loss /= len(train_loader.dataset)
    train_losses.append(train_loss)

    # Validation
    model.eval()
    val_loss = 0
    with torch.no_grad():
        for embeddings, mmse in val_loader:
            embeddings, mmse = embeddings.to(device), mmse.to(device)
            outputs = model(embeddings).squeeze()
            loss = criterion(outputs, mmse)
            val_loss += loss.item() * embeddings.size(0)
    val_loss /= len(val_loader.dataset)
    val_losses.append(val_loss)

    print(f"Epoch [{epoch+1}/{num_epochs}] Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}")

# -----------------------------
# 5️⃣ Plot Training vs Validation Loss
# -----------------------------
plt.figure(figsize=(8,5))
plt.plot(range(1, num_epochs+1), train_losses, label='Train Loss')
plt.plot(range(1, num_epochs+1), val_losses, label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('MSE Loss')
plt.title('Training vs Validation Loss')
plt.legend()
plt.show()

# -----------------------------
# 6️⃣ Evaluate on Test Set (Regression & Classification)
# -----------------------------
def mmse_to_class(mmse):
    """Convert MMSE score to dementia class"""
    if mmse < 12:
        return 0  # Severe
    elif 12 <= mmse <= 20:
        return 1  # Moderate
    elif 21 <= mmse <= 24:
        return 2  # Mild
    else:
        return 3  # Normal

model.eval()
preds, true = [], []

with torch.no_grad():
    for embeddings, mmse in test_loader:
        embeddings, mmse = embeddings.to(device), mmse.to(device)
        outputs = model(embeddings).squeeze()
        outputs = torch.clamp(outputs, 0, 30)  # Clip to 0-30
        preds.extend(outputs.cpu().numpy())
        true.extend(mmse.cpu().numpy())

# Regression Metrics
regression_mse = np.mean((np.array(true) - np.array(preds))**2)
regression_mae = np.mean(np.abs(np.array(true) - np.array(preds)))
print(f"\n📊 Regression Metrics:")
print(f"MSE: {regression_mse:.4f}, MAE: {regression_mae:.4f}")

# Convert to classes for classification metrics
true_classes = [mmse_to_class(x) for x in true]
pred_classes = [mmse_to_class(x) for x in preds]

accuracy = accuracy_score(true_classes, pred_classes)
precision = precision_score(true_classes, pred_classes, average='weighted', zero_division=0)
recall = recall_score(true_classes, pred_classes, average='weighted', zero_division=0)
f1 = f1_score(true_classes, pred_classes, average='weighted', zero_division=0)

print(f"\n📊 Classification Metrics (from MMSE ranges):")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")

# -----------------------------
# 7️⃣ Optional: Plot Predicted vs True MMSE (Regression)
# -----------------------------
plt.figure(figsize=(6,6))
plt.scatter(true, preds, color='blue', alpha=0.6)
plt.plot([0,30],[0,30], color='red', linestyle='--')
plt.xlabel('True MMSE')
plt.ylabel('Predicted MMSE')
plt.title('Predicted vs True MMSE')
plt.show()
```

```python
# ============================================================
# FINE-TUNE Data2Vec Audio Model + CNN Layer + Save CNN Embeddings
# ============================================================

import os
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, random_split
import librosa
import numpy as np
import pandas as pd
from transformers import AutoFeatureExtractor, Data2VecAudioModel
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix

# ============================================================
# 1️⃣ Dataset Class — Load Audio + Label
# ============================================================
class AlzheimerAudioDataset(Dataset):
    def __init__(self, csv_path, audio_dir, extractor, max_length=160000):
        self.data = pd.read_csv(csv_path)
        self.audio_dir = audio_dir
        self.extractor = extractor
        self.max_length = max_length

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        row = self.data.iloc[idx]
        file_path = os.path.join(self.audio_dir, row['file'])
        label = torch.tensor(row['label'], dtype=torch.float32).unsqueeze(0)
        
        speech_array, sr = librosa.load(file_path, sr=16000)
        if len(speech_array) < self.max_length:
            speech_array = np.pad(speech_array, (0, self.max_length - len(speech_array)))
        else:
            speech_array = speech_array[:self.max_length]

        inputs = self.extractor(
            speech_array,
            sampling_rate=sr,
            return_tensors="pt"
        )
        return inputs.input_values.squeeze(0), label, row['file']

# ============================================================
# 2️⃣ Model — Data2Vec + CNN + Classifier Head (save CNN emb)
# ============================================================
class Data2VecClassifier(nn.Module):
    def __init__(self, base_model_name="facebook/data2vec-audio-base-960h"):
        super().__init__()
        self.data2vec = Data2VecAudioModel.from_pretrained(base_model_name)

        # Freeze feature extractor
        for param in self.data2vec.feature_extractor.parameters():
            param.requires_grad = False

        # Freeze first 9 layers, unfreeze last 3
        for i, layer in enumerate(self.data2vec.encoder.layers):
            if i < 9:
                for param in layer.parameters():
                    param.requires_grad = False
            else:
                for param in layer.parameters():
                    param.requires_grad = True

        # CNN layer
        self.cnn = nn.Sequential(
            nn.Conv1d(in_channels=768, out_channels=350, kernel_size=3, padding=1),
            nn.BatchNorm1d(350),
            nn.ReLU(),
            nn.MaxPool1d(kernel_size=2)
        )

        # Classifier head
        self.classifier = nn.Sequential(
            nn.Linear(350, 300),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(300, 1),
            nn.Sigmoid()
        )

    def forward(self, input_values, save_emb_dir=None, file_names=None, save_cnn_emb=False):
        outputs = self.data2vec(input_values)
        hidden_states = outputs.last_hidden_state  # [batch, seq_len, 768]

        # CNN expects [batch, channels, seq_len]
        x = hidden_states.permute(0, 2, 1)  # [B, 768, seq_len]
        x = self.cnn(x)                     # [B, 350, seq_len/2]
        x_pooled = x.mean(dim=2)            # Global average pooling -> [B, 350]

        # Save CNN embeddings only
        if save_emb_dir is not None and file_names is not None and save_cnn_emb:
            os.makedirs(save_emb_dir, exist_ok=True)
            for i, fname in enumerate(file_names):
                emb = x_pooled[i].detach().cpu().numpy()
                np.save(os.path.join(save_emb_dir, fname.replace('.wav', '_cnn_embed.npy')), emb)

        logits = self.classifier(x_pooled)
        return logits

# ============================================================
# 3️⃣ Training + Evaluation
# ============================================================
def train_model(model, train_loader, val_loader, device, epochs=15, lr=1e-5, emb_save_dir=None):
    model.to(device)
    criterion = nn.BCELoss()
    optimizer = torch.optim.Adam(filter(lambda p: p.requires_grad, model.parameters()), lr=lr)

    train_losses, val_losses = [], []
    train_accs, val_accs = [], []

    for epoch in range(epochs):
        model.train()
        total_train_loss, correct_train, total_train = 0, 0, 0

        for x_batch, y_batch, file_names in train_loader:
            x_batch, y_batch = x_batch.to(device), y_batch.to(device)
            optimizer.zero_grad()
            outputs = model(x_batch, save_emb_dir=emb_save_dir, file_names=file_names, save_cnn_emb=True)
            loss = criterion(outputs, y_batch)
            loss.backward()
            optimizer.step()

            total_train_loss += loss.item() * x_batch.size(0)
            preds = (outputs > 0.5).float()
            correct_train += (preds == y_batch).sum().item()
            total_train += y_batch.size(0)

        avg_train_loss = total_train_loss / len(train_loader.dataset)
        avg_train_acc = correct_train / total_train

        # Validation
        model.eval()
        total_val_loss, correct_val, total_val = 0, 0, 0
        with torch.no_grad():
            for x_batch, y_batch, file_names in val_loader:
                x_batch, y_batch = x_batch.to(device), y_batch.to(device)
                outputs = model(x_batch)
                loss = criterion(outputs, y_batch)
                total_val_loss += loss.item() * x_batch.size(0)
                preds = (outputs > 0.5).float()
                correct_val += (preds == y_batch).sum().item()
                total_val += y_batch.size(0)

        avg_val_loss = total_val_loss / len(val_loader.dataset)
        avg_val_acc = correct_val / total_val

        train_losses.append(avg_train_loss)
        val_losses.append(avg_val_loss)
        train_accs.append(avg_train_acc)
        val_accs.append(avg_val_acc)

        print(f"Epoch {epoch+1}/{epochs} | "
              f"Train Loss: {avg_train_loss:.4f} | Val Loss: {avg_val_loss:.4f} | "
              f"Train Acc: {avg_train_acc*100:.2f}% | Val Acc: {avg_val_acc*100:.2f}%")

    # Plots
    plt.figure(figsize=(12,5))
    plt.subplot(1, 2, 1)
    plt.plot(train_losses, label='Train Loss')
    plt.plot(val_losses, label='Validation Loss')
    plt.xlabel("Epochs"); plt.ylabel("Loss"); plt.legend(); plt.title("Loss")

    plt.subplot(1, 2, 2)
    plt.plot(train_accs, label='Train Acc')
    plt.plot(val_accs, label='Val Acc')
    plt.xlabel("Epochs"); plt.ylabel("Accuracy"); plt.legend(); plt.title("Accuracy")

    plt.tight_layout()
    plt.show()

    return model

# ============================================================
# 4️⃣ Test Evaluation
# ============================================================
def evaluate_model(model, test_loader, device, emb_save_dir=None):
    model.eval()
    y_true, y_pred = [], []
    with torch.no_grad():
        for x_batch, y_batch, file_names in test_loader:
            x_batch = x_batch.to(device)
            outputs = model(x_batch, save_emb_dir=emb_save_dir, file_names=file_names, save_cnn_emb=True)
            preds = (outputs.cpu().numpy() > 0.5).astype(int)
            y_pred.extend(preds)
            y_true.extend(y_batch.cpu().numpy())

    acc = accuracy_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    print(f"\n✅ Test Accuracy: {acc:.4f} | F1 Score: {f1:.4f}")
    print("Confusion Matrix:\n", confusion_matrix(y_true, y_pred))


# ============================================================
# 5️⃣ Data Split
# ============================================================
csv_path = "/kaggle/input/alzhiemer/DATASET/label.csv"
audio_dir = "/kaggle/input/alzhiemer/DATASET"
extractor = AutoFeatureExtractor.from_pretrained("facebook/data2vec-audio-base-960h")

dataset = AlzheimerAudioDataset(csv_path, audio_dir, extractor)
train_size = int(0.8 * len(dataset))
val_size = int(0.1 * len(dataset))
test_size = len(dataset) - train_size - val_size
train_ds, val_ds, test_ds = random_split(dataset, [train_size, val_size, test_size])

train_loader = DataLoader(train_ds, batch_size=4, shuffle=True)
val_loader = DataLoader(val_ds, batch_size=4, shuffle=False)
test_loader = DataLoader(test_ds, batch_size=4, shuffle=False)

# ============================================================
# 6️⃣ Train + Evaluate + Save CNN Embeddings
# ============================================================
device = "cuda" if torch.cuda.is_available() else "cpu"
model = Data2VecClassifier()

embedding_folder = "/kaggle/working/cnn_embeddings"  # CNN embeddings only

model = train_model(model, train_loader, val_loader, device=device, epochs=12, lr=1e-5, emb_save_dir=embedding_folder)
evaluate_model(model, test_loader, device, emb_save_dir=embedding_folder)

print("✅ CNN embeddings saved only, raw Data2Vec embeddings are NOT saved.")
```

```python
def count_model_layers_simple(model):
    """Simple and accurate layer count for your specific model architecture"""
    
    # Initialize counters
    data2vec_frozen_layers = 0
    data2vec_trainable_layers = 0
    cnn_layers = 0
    classifier_layers = 0
    
    # Count layers by component
    for name, param in model.named_parameters():
        if 'data2vec' in name:
            if param.requires_grad:
                data2vec_trainable_layers += 1
            else:
                data2vec_frozen_layers += 1
        elif 'cnn' in name:
            cnn_layers += 1
        elif 'classifier' in name:
            classifier_layers += 1
    
    # Calculate totals
    total_layers = data2vec_frozen_layers + data2vec_trainable_layers + cnn_layers + classifier_layers
    total_trainable_layers = data2vec_trainable_layers + cnn_layers + classifier_layers
    total_frozen_layers = data2vec_frozen_layers
    
    print("=" * 60)
    print("MODEL LAYER BREAKDOWN")
    print("=" * 60)
    print(f"Data2Vec Audio Model:")
    print(f"  - Frozen layers: {data2vec_frozen_layers} (first 9 encoder layers + feature extractor)")
    print(f"  - Trainable layers: {data2vec_trainable_layers} (last 3 encoder layers)")
    print(f"  - Total Data2Vec layers: {data2vec_frozen_layers + data2vec_trainable_layers}")
    print(f"\nCNN Layers (added by you):")
    print(f"  - Trainable layers: {cnn_layers} (all CNN layers)")
    print(f"\nClassifier Head (added by you):")
    print(f"  - Trainable layers: {classifier_layers} (all classifier layers)")
    print(f"\nOVERALL SUMMARY:")
    print(f"  - Total layers: {total_layers}")
    print(f"  - Trainable layers: {total_trainable_layers}")
    print(f"  - Frozen layers: {total_frozen_layers}")
    print(f"  - Trainable ratio: {total_trainable_layers/total_layers*100:.1f}%")
    print("=" * 60)

# Add this function to count parameters as well
def count_parameters_simple(model):
    """Simple parameter count"""
    total_params = 0
    trainable_params = 0
    
    for param in model.parameters():
        total_params += param.numel()
        if param.requires_grad:
            trainable_params += param.numel()
    
    print("\n" + "=" * 60)
    print("PARAMETER COUNT")
    print("=" * 60)
    print(f"Total Parameters: {total_params:,}")
    print(f"Trainable Parameters: {trainable_params:,}")
    print(f"Frozen Parameters: {total_params - trainable_params:,}")
    print(f"Trainable Ratio: {trainable_params/total_params*100:.2f}%")
    print("=" * 60)

# Usage - add this after your model initialization
print("Analyzing your model architecture...")
count_model_layers_simple(model)
count_parameters_simple(model)
```

```python
# ============================================================
# ✅ EXTRACT CNN EMBEDDINGS FROM TRAIN + TEST + VALIDATION
# ✅ SAVE IN ONE FOLDER
# ✅ PRINT TOTAL FILES
# ============================================================

# ----------- 1) Create one main folder -------------
combined_emb_folder = "/kaggle/working/all_cnn_embeddings"
os.makedirs(combined_emb_folder, exist_ok=True)

# ----------- 2) Function to extract embeddings ------
def extract_embeddings(dataloader, model, device, save_folder):
    model.eval()
    saved_count = 0
    with torch.no_grad():
        for x_batch, y_batch, file_names in dataloader:
            x_batch = x_batch.to(device)

            # forward pass with save option
            _ = model(
                x_batch,
                save_emb_dir=save_folder,
                file_names=file_names,
                save_cnn_emb=True
            )
            saved_count += len(file_names)
    return saved_count

# ----------- 3) Run extraction for Train / Test / Validation -----

total_saved = 0

print("✅ Extracting TRAIN embeddings...")
total_saved += extract_embeddings(train_loader, model, device, combined_emb_folder)

print("✅ Extracting TEST embeddings...")
total_saved += extract_embeddings(test_loader, model, device, combined_emb_folder)

print("✅ Extracting VALIDATION embeddings...")
total_saved += extract_embeddings(val_loader, model, device, combined_emb_folder)

# ----------- 4) Print final summary ----------------
print("✅ ALL embeddings extracted!")
print(f"📌 Total embedding files saved: {total_saved}")
print(f"📁 Folder location: {combined_emb_folder}")
```

```python
# ============================================================
# 📌 Match CNN Embeddings with MMSE Scores and Save CSV
# ============================================================

import pandas as pd
import numpy as np
import os

# ============================================================
# 1️⃣ Load MMSE CSV properly
# ============================================================
mmse_csv_path = "/kaggle/input/mmse-label/mmse label.csv"
mmse_df = pd.read_csv(mmse_csv_path, header=None)

# Skip first row if it's header, take first 2 columns
mmse_df = mmse_df.iloc[1:, :2]  
mmse_df.columns = ['id', 'mmse']

# Clean up columns
mmse_df['id'] = mmse_df['id'].astype(str).str.strip()
mmse_df['mmse'] = pd.to_numeric(mmse_df['mmse'], errors='coerce')

print("✅ MMSE CSV loaded. Preview:")
print(mmse_df.head())

# ============================================================
# 2️⃣ Set the embeddings folder
# ============================================================
extract_dir = "/kaggle/input/cnn-embeded"

print("Files in the folder:")
print(os.listdir(extract_dir))

# ============================================================
# 3️⃣ Match embeddings with MMSE scores
# ============================================================
matched_data = []

for file_name in os.listdir(extract_dir):
    if file_name.endswith('.npy'):
        # Extract ID from filename (remove suffixes like _cnn_embed)
        file_id = file_name.replace('.npy','').replace('_cnn_embed','')  

        # Check if ID exists in MMSE CSV
        if file_id in mmse_df['id'].values:
            mmse_score = mmse_df.loc[mmse_df['id'] == file_id, 'mmse'].values[0]
            embedding_data = np.load(os.path.join(extract_dir, file_name))
            
            matched_data.append({
                'file_name': file_name,
                'id': file_id,
                'mmse_score': mmse_score,
                'embedding_shape': embedding_data.shape,
            })
            print(f"✅ Matched: {file_name} -> ID: {file_id}, MMSE: {mmse_score}")
        else:
            print(f"❌ No MMSE score found for: {file_id}")

# ============================================================
# 4️⃣ Create DataFrame and save
# ============================================================
matched_df = pd.DataFrame(matched_data)
output_csv = "/kaggle/working/matched_embeddings_mmse.csv"
matched_df.to_csv(output_csv, index=False)

# ============================================================
# 5️⃣ Summary
# ============================================================
print(f"\n📊 Matching Summary:")
print(f"Total embedding files: {len(os.listdir(extract_dir))}")
print(f"Successfully matched: {len(matched_df)}")
print(f"Unmatched files: {len(os.listdir(extract_dir)) - len(matched_df)}")
print(f"💾 Matched embeddings + MMSE saved to: {output_csv}")
```

```python
# ============================================================
# MMSE Prediction from CNN Embeddings (Complete Working Code)
# ============================================================

import os
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# -----------------------------
# 1️⃣ Paths (FIXED)
# -----------------------------
csv_path = "/kaggle/input/mmse-label/mmse_labels.csv"   # CSV file
embeddings_dir = "/kaggle/input/cnn-embeded"           # Folder with .npy files

df = pd.read_csv(csv_path)

# CSV must contain:
# file_name | mmse_score

# -----------------------------
# 2️⃣ Dataset
# -----------------------------
class EmbeddingDataset(Dataset):
    def __init__(self, df, embeddings_dir):
        self.df = df.reset_index(drop=True)
        self.embeddings_dir = embeddings_dir

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        emb = np.load(os.path.join(self.embeddings_dir, row["file_name"]))
        emb = torch.tensor(emb, dtype=torch.float32).squeeze()
        mmse = torch.tensor(row["mmse_score"], dtype=torch.float32)
        return emb, mmse

dataset = EmbeddingDataset(df, embeddings_dir)

# -----------------------------
# 3️⃣ Train / Val / Test Split
# -----------------------------
train_size = int(0.8 * len(dataset))
val_size = int(0.1 * len(dataset))
test_size = len(dataset) - train_size - val_size

train_set, val_set, test_set = random_split(dataset, [train_size, val_size, test_size])

train_loader = DataLoader(train_set, batch_size=8, shuffle=True)
val_loader = DataLoader(val_set, batch_size=8)
test_loader = DataLoader(test_set, batch_size=8)

# -----------------------------
# 4️⃣ Model
# -----------------------------
class ADSeverityNN(nn.Module):
    def __init__(self, input_dim=350):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, 300),
            nn.ReLU(),
            nn.Linear(300, 150),
            nn.ReLU(),
            nn.Dropout(0.1),
            nn.Linear(150, 1)
        )

    def forward(self, x):
        return self.net(x)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = ADSeverityNN(input_dim=350).to(device)

criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)

# -----------------------------
# 5️⃣ Training Loop
# -----------------------------
num_epochs = 50
train_losses, val_losses = [], []

for epoch in range(num_epochs):
    model.train()
    train_loss = 0

    for x, y in train_loader:
        x, y = x.to(device), y.to(device)
        optimizer.zero_grad()
        preds = model(x).view(-1)
        loss = criterion(preds, y)
        loss.backward()
        optimizer.step()
        train_loss += loss.item() * x.size(0)

    train_loss /= len(train_loader.dataset)
    train_losses.append(train_loss)

    model.eval()
    val_loss = 0
    with torch.no_grad():
        for x, y in val_loader:
            x, y = x.to(device), y.to(device)
            preds = model(x).view(-1)
            loss = criterion(preds, y)
            val_loss += loss.item() * x.size(0)

    val_loss /= len(val_loader.dataset)
    val_losses.append(val_loss)

    print(f"Epoch {epoch+1}/{num_epochs} | Train: {train_loss:.4f} | Val: {val_loss:.4f}")

# -----------------------------
# 6️⃣ Normalized Epoch Loss Plot (BEST for PPT)
# -----------------------------
epochs_norm = np.linspace(0, 1, num_epochs)

plt.figure(figsize=(7,5))
plt.plot(epochs_norm, train_losses, label="Train Loss")
plt.plot(epochs_norm, val_losses, label="Validation Loss")
plt.xlabel("Normalized Training Progress")
plt.ylabel("MSE Loss")
plt.title("Training vs Validation Loss")
plt.legend()
plt.grid(True)
plt.show()

# -----------------------------
# 7️⃣ Test Evaluation
# -----------------------------
def mmse_to_class(mmse):
    if mmse < 12:
        return 0  # Severe
    elif 12 <= mmse <= 20:
        return 1  # Moderate
    elif 21 <= mmse <= 24:
        return 2  # Mild
    else:
        return 3  # Normal

model.eval()
preds, true = [], []

with torch.no_grad():
    for x, y in test_loader:
        x, y = x.to(device), y.to(device)
        out = model(x).view(-1)
        out = torch.clamp(out, 0, 30)
        preds.extend(out.cpu().numpy())
        true.extend(y.cpu().numpy())

# Regression metrics
true = np.array(true)
preds = np.array(preds)

mse = np.mean((true - preds) ** 2)
mae = np.mean(np.abs(true - preds))

print("\n📊 Regression Metrics")
print(f"MSE: {mse:.4f}")
print(f"MAE: {mae:.4f}")

# Classification metrics
true_cls = [mmse_to_class(x) for x in true]
pred_cls = [mmse_to_class(x) for x in preds]

acc = accuracy_score(true_cls, pred_cls)
prec = precision_score(true_cls, pred_cls, average="weighted", zero_division=0)
rec = recall_score(true_cls, pred_cls, average="weighted", zero_division=0)
f1 = f1_score(true_cls, pred_cls, average="weighted", zero_division=0)

print("\n📊 Classification Metrics (MMSE ranges)")
print(f"Accuracy : {acc:.4f}")
print(f"Precision: {prec:.4f}")
print(f"Recall   : {rec:.4f}")
print(f"F1 Score : {f1:.4f}")

# -----------------------------
# 8️⃣ Predicted vs True MMSE Plot
# -----------------------------
plt.figure(figsize=(6,6))
plt.scatter(true, preds, alpha=0.6)
plt.plot([0,30], [0,30], "r--")
plt.xlabel("True MMSE")
plt.ylabel("Predicted MMSE")
plt.title("Predicted vs True MMSE")
plt.grid(True)
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# -----------------------------
# 1️⃣ Load matched CSV and embeddings
# -----------------------------
csv_path = "/kaggle/input/cnn-embeded"
embeddings_dir = "/kaggle/input/cnn-embeded"
df = pd.read_csv(csv_path)

# Custom Dataset
class EmbeddingDataset(Dataset):
    def __init__(self, df, embeddings_dir):
        self.df = df
        self.embeddings_dir = embeddings_dir

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        embedding = np.load(os.path.join(self.embeddings_dir, row['file_name']))
        embedding = torch.tensor(embedding, dtype=torch.float32).squeeze()
        mmse = torch.tensor(row['mmse_score'], dtype=torch.float32)
        return embedding, mmse

dataset = EmbeddingDataset(df, embeddings_dir)

# -----------------------------
# 2️⃣ Train / Val / Test Split
# -----------------------------
total = len(dataset)
train_size = int(0.8 * total)
val_size = int(0.1 * total)
test_size = total - train_size - val_size

train_set, val_set, test_set = random_split(dataset, [train_size, val_size, test_size])
train_loader = DataLoader(train_set, batch_size=4, shuffle=True)
val_loader = DataLoader(val_set, batch_size=4)
test_loader = DataLoader(test_set, batch_size=4)

# -----------------------------
# 3️⃣ Neural Network
# -----------------------------
class ADSeverityNN(nn.Module):
    def __init__(self, input_dim=350):
        super(ADSeverityNN, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(input_dim, 300),
            nn.ReLU(),
            nn.Linear(300, 150),
            nn.Dropout(0.1),
            nn.ReLU(),
            nn.Linear(150, 1)
        )

    def forward(self, x):
        return self.model(x)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = ADSeverityNN(input_dim=350).to(device)

# Loss and optimizer
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)

# -----------------------------
# 4️⃣ Training Loop with Loss History
# -----------------------------
num_epochs = 50
train_losses, val_losses = [], []

for epoch in range(num_epochs):
    # Training
    model.train()
    train_loss = 0
    for embeddings, mmse in train_loader:
        embeddings, mmse = embeddings.to(device), mmse.to(device)
        optimizer.zero_grad()
        outputs = model(embeddings).view(-1)  # UPDATED here
        loss = criterion(outputs, mmse)
        loss.backward()
        optimizer.step()
        train_loss += loss.item() * embeddings.size(0)
    train_loss /= len(train_loader.dataset)
    train_losses.append(train_loss)

    # Validation
    model.eval()
    val_loss = 0
    with torch.no_grad():
        for embeddings, mmse in val_loader:
            embeddings, mmse = embeddings.to(device), mmse.to(device)
            outputs = model(embeddings).view(-1)  # UPDATED here
            loss = criterion(outputs, mmse)
            val_loss += loss.item() * embeddings.size(0)
    val_loss /= len(val_loader.dataset)
    val_losses.append(val_loss)

    print(f"Epoch [{epoch+1}/{num_epochs}] Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}")

# -----------------------------
# 5️⃣ Plot Training vs Validation Loss
# -----------------------------
plt.figure(figsize=(8,5))
plt.plot(range(1, num_epochs+1), train_losses, label='Train Loss')
plt.plot(range(1, num_epochs+1), val_losses, label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('MSE Loss')
plt.title('Training vs Validation Loss')
plt.legend()
plt.show()

# -----------------------------
# 6️⃣ Evaluate on Test Set (Regression & Classification)
# -----------------------------
def mmse_to_class(mmse):
    """Convert MMSE score to dementia class"""
    if mmse < 12:
        return 0  # Severe
    elif 12 <= mmse <= 20:
        return 1  # Moderate
    elif 21 <= mmse <= 24:
        return 2  # Mild
    else:
        return 3  # Normal

model.eval()
preds, true = [], []

with torch.no_grad():
    for embeddings, mmse in test_loader:
        embeddings, mmse = embeddings.to(device), mmse.to(device)
        outputs = model(embeddings).view(-1)  # UPDATED here
        outputs = torch.clamp(outputs, 0, 30)  # Clip to 0-30
        preds.extend(outputs.cpu().numpy())
        true.extend(mmse.cpu().numpy())

# Regression Metrics
regression_mse = np.mean((np.array(true) - np.array(preds))**2)
regression_mae = np.mean(np.abs(np.array(true) - np.array(preds)))
print(f"\n📊 Regression Metrics:")
print(f"MSE: {regression_mse:.4f}, MAE: {regression_mae:.4f}")

# Classification Metrics
true_classes = [mmse_to_class(x) for x in true]
pred_classes = [mmse_to_class(x) for x in preds]

accuracy = accuracy_score(true_classes, pred_classes)
precision = precision_score(true_classes, pred_classes, average='weighted', zero_division=0)
recall = recall_score(true_classes, pred_classes, average='weighted', zero_division=0)
f1 = f1_score(true_classes, pred_classes, average='weighted', zero_division=0)

print(f"\n📊 Classification Metrics (from MMSE ranges):")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")

# -----------------------------
# 7️⃣ Plot Predicted vs True MMSE
# -----------------------------
plt.figure(figsize=(6,6))
plt.scatter(true, preds, color='blue', alpha=0.6)
plt.plot([0,30],[0,30], color='red', linestyle='--')
plt.xlabel('True MMSE')
plt.ylabel('Predicted MMSE')
plt.title('Predicted vs True MMSE')
plt.show()
```
